##### POS Tagging Pipeline

An example of the Part-Of-Speech (**POS**) tagging and parsing as performed using Spacy's models. 

This is the process that happens during the testing loop if the **automatic_pos_tagging** variable is set to **TRUE** to tag the words in each sentence with their corresponding POS tag and therefore to check and evaluate if this process has a positive performance on the overall named entity recognition task or not.

```python
import spacy
from spacy import displacy

model_name = "en_core_web_md"
pos_tagger = spacy.load(model_name, disable = ['ner', 'lemmatizer'])
print('POS Tagging Pipe Lines: {}'.format(pos_tagger.pipe_names))


example_sentence = "In order to support cultural businesses and creative artists, the Federal Office of Culture (FOC) may enter into service agreements with one or more cantons."
doc = pos_tagger(example_sentence)

##### The example sentence is tokenized and then Universal POS Tags are saved into it's token.pos_ variable
# for i, token in enumerate(doc):
#         print(token.text, token.pos_)

##### For a better visualization in Jupyter, Displacy is again used
displacy.render(doc, style = 'dep')
```
